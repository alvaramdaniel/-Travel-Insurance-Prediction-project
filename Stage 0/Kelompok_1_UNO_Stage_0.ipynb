```python
# Melihat data head
df.head()
```

```python
# Menghapus unamed kolom yang tidak dibutuhkan
df.drop(columns=['Unnamed: 0'],inplace=True)
df.head()
```

```python
# Mengecek dataset info
df.info()
```

**Dataset tidak memiliki null value di tiap kolomnya dan semua kolom memiliki tipe data yang sesuai.**

```python
mapping_employmenttype = {
    'Government Sector' : 0,
    'Private Sector/Self Employed' : 1
}

df['Employment Type'] = df['Employment Type'].map(mapping_employmenttype) 

df.head()
```

```python
mapping_graduate = {
    'Yes' : 1,
    'No' : 0
}

df['GraduateOrNot'] = df['GraduateOrNot'].map(mapping_graduate) 

df.head()
```

```python
mapping_freqflyer = {
    'Yes' : 1,
    'No' : 0
}

df['FrequentFlyer'] = df['FrequentFlyer'].map(mapping_freqflyer) 

df.head()
```

```python
mapping_travabr = {
    'Yes' : 1,
    'No' : 0
}

df['EverTravelledAbroad'] = df['EverTravelledAbroad'].map(mapping_travabr) 

df.head()
```
